```python
# Step 1: Install OpenCV
!pip install opencv-python-headless

# Step 2: Import necessary libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Function to display images using Matplotlib
def display_image(img, title="Image"):
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.axis('off')
    plt.show()

# Step 3: Load the image (you can replace 'image.jpg' with your own image)
# Upload the image in Google Colab manually or use any online image URL to read
from google.colab import files
uploaded = files.upload()

# Load the image
image_path = list(uploaded.keys())[0]
image = cv2.imread(image_path)

# Step 4: Define the smoothing filter (3x3 kernel mask)
# A basic averaging filter mask
kernel = np.array([[1/9, 1/9, 1/9],
                   [1/9, 1/9, 1/9],
                   [1/9, 1/9, 1/9]])

# Step 5: Apply the mask using cv2.filter2D
smoothed_image = cv2.filter2D(image, -1, kernel)

# Step 6: Display the original and smoothed images
plt.figure(figsize=(10,5))

plt.subplot(1, 2, 1)
display_image(image, "Original Image")

plt.subplot(1, 2, 2)
display_image(smoothed_image, "Smoothed Image (Smoothing Filter)")

# Step 7: Save the result (optional)
cv2.imwrite('smoothed_image.jpg', smoothed_image)
```
